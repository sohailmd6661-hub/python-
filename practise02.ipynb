```python
class GK:
  a=200
  b=40
  def addition(self):
    print(self.a+self.b)

  def subtraction(self):
    print(self.a-self.b)

  def multiply(self):
    print(self.a * self.b)

  def division(self):
    print(self.a / self.b)

anil = GK()
anil.addition()
anil.subtraction()
anil.multiply()
anil.division()
```

```python
class GK:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

  def multiply(self):
    print(self.a * self.b)

anil = GK(10,20)
anil.addition()
anil.subtraction()
anil.multiply()
```

```python
# with constructor

class CALCULATOR:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

  def multiply(self):
    print(self.a * self.b)

  def division(self):
    print(self.a / self.b)

anil = CALCULATOR(178,289)
anil.addition()
anil.subtraction()
anil.multiply()
anil.division()
```
