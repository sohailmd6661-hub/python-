```python
# class = it is the place where varibales and functions are stored
# object = incharge of a class

class CSE:
  john = 20
  peter = 30

anil = CSE() # anil is an object
print(anil.john)
```

```python
class ECE:
  f = 30
  g = 80

kumar = ECE()
print(kumar.f * kumar.g)
```

```python
# functions in class

class CSE:
  f = 100

  def fun():
    print('GM')


anil = CSE()
print(anil.f)
```

```python
# functions in class

class CSE:
  f = 100

  def sharuk(b):   # b is a varibale which is holding object (anil) copy
    print(b)
    print('GM')


anil = CSE()
print(anil)
print(anil.f)
anil.sharuk()
```

```python
# functions in class

class CSE:
  f = 100

  def sharuk(self):   # self is a varibale which is holding object (anil) copy
    print('GM')


anil = CSE()
print(anil)
print(anil.f)
anil.sharuk()
```

```python
class MECH:
  r = 200

  def kumar():
    print('GM')
    print(f'Hi : {r}')

prem = MECH()
print(prem.r)
```

```python
class MECH:
  r = 200

  def kumar(self): # self is varibale holding (prem)
    print('GM')
    print(f'Hi : {self.r}')

prem = MECH()
prem.kumar()
```

```python
class CSE:
  a = 10
  b = 20

  def addition(self):
    print(self.a + self.b)

  def subtract(self):
    print(self.a - self.b)

  def common(self):
    self.addition()
    self.subtract()


anil = CSE()
anil.common()
```
