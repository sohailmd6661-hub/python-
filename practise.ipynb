Exception Practise

```python
try:
  a = 0 # save in the Memory
  b = 20 # save in the Memory
  print(b / a) # 8lines


except Exception as e:
  print(e)
```

```python
try:
  a = 10
  b = 0
  print(a / b)

except ValueError as e:
  print(e)

except NameError as e:
  print(e)
```

```python
try:
    x = 10
    y = 2
    print(x / y)
except ZeroDivisionError:
    print("Error")
else:
    print("No exception occurred")
```

```python
try:
    x = int("10")
except ValueError:
    print("Error")
finally:
    print("Execution completed")
```

```python
balance = 5000
try:
    withdraw = int(input("Enter amount: "))
    if withdraw > balance:
        raise ValueError("Insufficient balance")
    balance -= withdraw
    print("Collect cash")
except ValueError as e:
    print(e)
```
