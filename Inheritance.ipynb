**Inheritance**  : parent and child relation

- `Single Inheritance `
- `Multi Level Inheritance `
- `Multiple Inheritance `
- `Hierarichal Inheritance `
- `Hybrid Inheritance `

```python
# single Inheritance

class PARENT:
  def building(self):
    print(f'I have My own Buildng')


class CHILD(PARENT):
  def job(self):
    print(f'I have my own Job')


sohail = CHILD()
sohail.job()
sohail.building()
```

```python
# single Inheritance

class PARENT:
  def building(self):
    print(f'I have My own Buildng')


class CHILD(PARENT):
  def job(self):
    print(f'I have my own Job')


sohail = CHILD()
sohail.job()
sohail.building()
```

```python
# Multi Level Inheritance

class GRANDFATHER:
  def shopping_mall(self):
    print(f'I have My own Shopping Mall')

class FATHER(GRANDFATHER):
  def building(self):
    print(f'I have My own Buildng')


class CHILD(FATHER):
  def job(self):
    print(f'I have my own Job')

sohail = CHILD()
sohail.job()
sohail.building()
sohail.shopping_mall()
```

```python
# Multiple Inheritance

class MOTHER:
  def shopping_mall(self):
    print(f'I have My own Shopping Mall')

class FATHER:
  def building(self):
    print(f'I have My own Buildng')


class CHILD(FATHER,MOTHER):
  def job(self):
    print(f'I have my own Job')

sohail = CHILD()
sohail.job()
sohail.building()
sohail.shopping_mall()
```

```python
# Hierarichal Inhertiance

class FATHER:
  def shopping_mall(self):
    print(f'I have My own Shopping Mall')

class B1(FATHER):
  def building(self):
    print(f'I have My own Buildng')


class B2(FATHER):
  def job(self):
    print(f'I have my own Job')

class B3(FATHER):
  def PROPERTY_(self):
    print(f'I have my own RICH PROPERTY')


sohail = B2()
sohail.job()
sohail.shopping_mall()
```

```python
# hybrid inheritance

class A:
    def showA(self):
        print("Class A")

class B(A):        # Single inheritance
    def showB(self):
        print("Class B")

class C(A):        # Single inheritance
    def showC(self):
        print("Class C")

class D(B, C):     # Multiple inheritance (Hybrid)
    def showD(self):
        print("Class D")

obj = D()
obj.showA()
obj.showB()
obj.showC()
obj.showD()
```
