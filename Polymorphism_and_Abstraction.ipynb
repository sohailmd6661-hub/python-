Abstraction


```python
# Abstraction : Hiding internal details and showing only essential features of an object.


from abc import ABC,abstractmethod


class WHATSAPP_VERSION_1(ABC):
  def chat(self):
    print('chat available')

  @abstractmethod
  def status(self):
    pass


class WHATSPP_VERSION_2(WHATSAPP_VERSION_1):
  def status(self):
    print(f'New changes ')


obj = WHATSPP_VERSION_2()
obj.status()
```
