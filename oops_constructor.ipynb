**Constructor**

`Constructor is used for initalizing what we needed (Class name and Constructor name will be same)`

`There are 2 types of Constructors `
    
          - `Non_Parameter Constructors`
          - `Parameter Constructors`

```python
class MECH:
  f = 20

  def __init__(self):
    print('Good Morning')

anil = MECH()
```

```python
class MECH:
  f = 20

  def __init__(self,a):
    print('Good Morning',a)

anil = MECH(10)
```

```python
class CSE:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

anil = CSE(6,8)
anil.addition()
anil.subtraction()
```

```python
# without constructor

class CALCULATOR:
  a = 6
  b = 8
  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

  def multiply(self):
    print(self.a * self.b)

  def division(self):
    print(self.a / self.b)

anil = CALCULATOR()
anil.addition()
anil.subtraction()
anil.multiply()
anil.division()
```

```python
# with constructor

class CALCULATOR:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

  def multiply(self):
    print(self.a * self.b)

  def division(self):
    print(self.a / self.b)

anil = CALCULATOR(60,80)
anil.addition()
anil.subtraction()
anil.multiply()
anil.division()
```

```python
class PRIME:
  def calculation(self,a,b):
    for i in range(a,b): # (1,101)
      c = 0
      for j in range(1,i+1):
        if i % j == 0:
          c = c + 1
      if c == 2:
        print(f'Prime : {i}')
      else:
        print(f'Not Prime : {i}')

anil = PRIME()
anil.calculation(1,101)
```
