**Built in function which is use in  pandas for AI**

**series** :
We can create only 1 column.

**Data frame** :
We can create proper excel data like ...dictionary key_value.

**Shape** :
 From this we can know how many rows and column.

 **Size** :
We can know how many count of total Dimension.


**Loc** :
We will not use (n-1) and no index.

**Iloc** :
we will use (n-1) and index.

**Rename** :
Using this we can give another name.

**Drop** :
Its means remove/delete rows and columns

**Axis = 1**  :
Columns

**Axis = 0**:
Rows

**to_csv** :
"comma seperate valve" save a Data frame as a csv file.

**to_read** :
Reads a csv file into a pandas Dataframe.

**Head** :
Its show top 5 rows of Dataframe.

**Tail** :
Its show last 5 rows of Dataframe.

**Sample** :
Return random rows from a Dataframe.

**isnull().sum()** :
Counts missing (NaN) values in each column.

**info** :
Get all information from Dataframe.

**unique** :
Return unique values from a columns.

**value_counts** :
Counts frequency of each unique value.

**select_dtypes** :
Selects columns of a specific data type from a Data frame.

**group  by ** :
Groups data by a column to perform calculation.

**fill na** :
Fills missing (NaN) values with a specified value.

**drop na** :
Remove rows and columns with missing (NaN) values.

**index** :
Label of rows in a Data frame.

**columns** :
Label of data field in a data frame.
