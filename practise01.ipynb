```python
class Car:
    def start(self):
        print("Car is starting")

    def stop(self):
        print("Car is stopped")

# object creation
c1 = Car()
c1.start()
c1.stop()
```

```python
class ATM:
    def __init__(self, pin):
        self.__pin = pin   # private variable

    def show_pin(self):
        print("Your PIN is protected")

atm = ATM(1234)
atm.show_pin()


```

```python
class Student:
    def __init__(self):
        print("Student object created")

s1 = Student()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Sohail", 22)
print(s1.name)
print(s1.age)
```

```python
class CAR:
    def __init__(self, brand="Honda"):
        self.brand = brand

c1 = CAR()          # non-parameterized
c2 = CAR("BMW")     # parameterized

print(c1.brand)
print(c2.brand)
```
