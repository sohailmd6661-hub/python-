**Types of Blocks**
1. try

Code that may cause an error

2. except

Handles the error

3. else

Executes if no exception occurs

4. finally

Executes always (cleanup code)

```python
n= 5
s = 0

try:
    for i in range(1, n):
        if n % i == 0:
            s = s + i

    if s == n:
        print(f'{n} : is a Perfect Number')
    else:
        print(f'{n} : is not a Perfect Number')
except Exception as e:
    print("An error occurred:", e)
finally:
    print("Done")
```
